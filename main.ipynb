Import Libraries

```python
import numpy as np
import pandas as pd
import os
from ultralytics import YOLO
import cv2
import pandas as pd
import numpy as np
import PIL
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
import os
import pathlib
```

Import Images

```python
for dirname, _, filenames in os.walk('/content/drive/MyDrive/object detection'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Generation of model

```python
model = YOLO("yolov8m.pt")
```

Storing the model into the variable callded results

```python
results=model.predict(source="/content/drive/MyDrive/object detection/images/1478019953180167674.jpg",
              save=True, conf=0.2,iou=0.5)
```

```python
result = results[0]
```

```python
Plotting the Detection of Objects
```

```python
res_plotted = results[0].plot()
res_plotted = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)
display(Image.fromarray(res_plotted))
```
